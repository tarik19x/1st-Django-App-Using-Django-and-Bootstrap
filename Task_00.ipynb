##$$\color{blue}{\textbf{Signal-01: Signals  and  Plotting  the  signal.} }$$
<br>


**Signals** are the electric or electromagnetic impulses used to encode and transmit data.In data communications, periodic analog signals and nonperiodic digital signals are are commonly used. 
 periodic analog signal is composed of multiple sine waves. non periodic digital signal is used in representing different levels of signals in a chanel 

```python
# reg. no: 2017331019
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

A sine wave or sinusoidal wave is the most natural representation of how many things in nature change state. A sine wave shows how the amplitude of a variable changes with time.

$$\color{black}{\textbf{The formula of Sinusoidal Signal  x(t)} }$$
$$\color{#752408}{x(t) = A\sin(2\pi f t + phase)}$$

```python
t = np.linspace(0, 2.0, 120)
def create_sinusoidal(A, f, t, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x
```

```python
x = create_sinusoidal(1,1,t)
sns.set()

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 12))
fig.suptitle('\nDifferent subplots of a sinusoidal signal', fontsize=15, color="red")

ax[0, 0].plot(t,x)
ax[0, 0].set_title('Continuous plot', fontsize = 15)

ax[0, 1].stem(t,x)
ax[0, 1].set_title('Stem plot', fontsize = 15)

ax[1,0].scatter(t,x)
ax[1,0].set_title('Scatter plot', fontsize = 15)
ax[-1,-1].axis('off')
for a in ax.flat:
    a.set(xlabel='time(t)', ylabel='x(t)')

plt.show()
```

<br>
<br>

When we change one or more characteristics of a single, single--frequency signal, it becomes a **Composite signal** made of many frequencies. A composite signal can be periodic or non-periodic. Composite signal is a combination of simple sine waves with different frequencies, amplitudes, and phases.A single-frequency sine wave is not useful in data communications;we need to send a composite signal, a signal made of many simple sine waves. here **y(t)** is a **Composite signal**

<br>
<br>

```python
x1 = create_sinusoidal(1,2,t)
x2 = create_sinusoidal(2,3,t)
x3 = create_sinusoidal(3,5,t)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 12))
fig.suptitle('3 Sinusoidal signal and a composite signal', fontsize=15, color="red")

ax[0, 0].plot(t,x1, 'r')
ax[0, 0].set_title('x1(t)', fontsize = 15)

ax[0, 1].plot(t,x2, 'g')
ax[0, 1].set_title('x2(t)', fontsize = 15)

ax[1,0].plot(t,x3, 'b')
ax[1,0].set_title('x3(t)', fontsize = 15)

ax[1,1].plot(t, x1 + x2 + x3)
ax[1,1].set_title('y(t) = x1(t) + x2(t) + x3(t)', fontsize = 15)
plt.show()
```

###$$\color{#752408}{A\sin(2\pi f t + phase)=x(t)  \to x(n)}$$

<br>
Sampling should be performed in order to turn a continuous-time signal into a discrete-time signal. The sample rate or frequency should be at least twice as high as the original frequency, if not higher.
<br>

```python
fs = 20
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(1,1,n)
plt.title('fs = 10', fontsize = 15)
plt.stem (n,y)
plt.show()
```

$$\color{green}{\textbf{Propert-01: Shifting of a signal} }$$

Shifting means movement of the signal, either in time domain aroundYaxis or in amplitude domain aroundXaxis. Accordingly, we can classify the shifting into two categories named as Time shifting and Amplitude shifting, these are subsequently discussed below.
A signal <font color='blue'>x(n)</font> may be shifted
in time by replacing the independent variable n by (nk), where k is an integer. If k
is a positive integer, the time shift results in a <font color='red'>**delay**</font>  of the signal by k units of time.
If k is a negative integer, the time shift results in an <font color='red'>**advance**</font> of the signal by |k| units
in time.

```python
def Shifting(y,k,n):
  ret=np.zeros(abs(k))

  if(k>=0):
    ret=np.concatenate((ret,y[0:len(n)-k]))
  else:
    ret=np.concatenate((y[abs(k):len(n)],ret))
  return ret
```

```python
k=7
plt.stem(n,Shifting(y,k,n))
plt.title('Delay', fontsize = 15)
plt.show()

k=-5
plt.stem(n,Shifting(y,k,n))
plt.title('Advance', fontsize = 15)
plt.show()
```

$$\color{green}{\textbf{Propert-02: Folding or Reflection} }$$

If the independent variable **t** is replaced by **t** , this operation is known as time reversal of the signal about the y-axis or amplitude axis. Hence, time reversal is known as folding or reflection. Folding or Reflection of the signal about the time origin n = 0 is the time base replacement of the independent variable n by n.

```python
plt.stem(n,y)
plt.title("Orginal Signal")
plt.show()

plt.stem(-n,y)
plt.title("Folded/Reflected Signal")
plt.show()
```
